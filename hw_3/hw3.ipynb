Самое популярное мужское имя

```python

Name=df.Name.str.extract('^.+?\w+\..+?\((.+?)\)|^.+?\w+\.\s([\w\s]+)',expand=False)
Name.fillna('',inplace=True)
Name.columns = ['A','B']
Name['CombinedName']=Name['A']+Name['B']
Name['Sex']=df['Sex']
Name['Class']=df['Pclass']
Name1 = Name[Name.Sex == 'male']
MaleName=Name1['CombinedName']

def pop_name (Name):
  menNames=[]
  for i in range(Name.count()-1):
    names=Name.iloc[i+1]
    words=names.split(' ')
    for word in words:
        if len(word)>2 and word.find('"'):
            menNames.append(word)

  MenNamesDataFrame=pd.DataFrame(data=menNames,columns=['Names'])
  MenNamesDataFrame['count']= 1
  MenNamesDataFrameGroupBy=MenNamesDataFrame.groupby('Names').sum()
  MenNamesDataFrameSorted=MenNamesDataFrameGroupBy.sort_values(['count'], ascending=[False])
  print(MenNamesDataFrameSorted.head(1))

pop_name(MaleName)
```
