```python
#@title Books
import requests
import ast
import pandas as pd
api_base_url='https://anapioficeandfire.com/api/'
books = requests.get(f'{api_base_url}/books/1').text
s=pd.Series(ast.literal_eval(books))
df_books=s.to_frame().T
i=2
while requests.get(f'{api_base_url}/books/{i}').text!='':
  books2 = requests.get(f'{api_base_url}/books/{i}').text
  s2=pd.Series(ast.literal_eval(books2))
  df_books2=s2.to_frame().T
  df_books=pd.concat([df_books, df_books2], ignore_index=True)
  i=i+1

df_books
```

```python
#@title Houses
houses_list = requests.get(f'{api_base_url}/houses', ).json()
df_houses=pd.DataFrame(houses_list)
i=2
while  requests.get(f'{api_base_url}/houses?page={i}').text!='' and df_houses.shape[0]%10==0:
  houses_list2 = requests.get(f'{api_base_url}/houses?page={i}', ).json()
  df_houses2=pd.DataFrame(houses_list2)
  df_houses=pd.concat([df_houses,df_houses2], ignore_index=True)
  i=i+1

df_houses
```

```python
#@title Houses with words
houses_list = requests.get(f'{api_base_url}/houses?hasWords=true', ).json()
df_houses=pd.DataFrame(houses_list)
i=2
while  requests.get(f'{api_base_url}/houses?hasWords=true&page={i}').text!='' and df_houses.shape[0]%10==0:
  houses_list2 = requests.get(f'{api_base_url}/houses?hasWords=true&page={i}', ).json()
  df_houses2=pd.DataFrame(houses_list2)
  df_houses=pd.concat([df_houses,df_houses2], ignore_index=True)
  i=i+1

df_houses
```
