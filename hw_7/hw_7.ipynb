```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from tqdm import tqdm

base_url = 'https://lifehacker.ru/topics/technology/'
response = requests.get(base_url)
soup = BeautifulSoup(response.text, 'lxml')
raw_items = soup.find_all('a', class_='lh-article-card-controls__comments')
parsed_urls = [item.get('href').replace('/#comments', '') for item in raw_items]
parsed_urls
```

```python
base_url = 'https://lifehacker.ru'
result = []
for url in tqdm(parsed_urls[:10]):
    article = {}
    article_url = f'{base_url}{url}'
    response = requests.get(article_url)
    soup = BeautifulSoup(response.text, 'lxml')

    article['title'] = soup.find('h1', class_='article-card__title').text
    text=soup.find('div', class_='single-article__post-content').text
    textReadAlso=soup.find('div', class_='read-also').text
    if(soup.find('div', class_='table-of-contents__menu'))!=None:
       textContent=(soup.find('div', class_='table-of-contents__menu').text)

    article['text'] = text.replace(textReadAlso, '').replace(textContent, '')
    result.append(article)

pd.set_option('display.max_colwidth', 700)
data = pd.DataFrame(result)
data

```
