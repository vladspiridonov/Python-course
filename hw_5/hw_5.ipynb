```python
#@title Основная информация
#!pip install bottleneck
import pandas as pd
import polars as pl
import numpy as np
import bottleneck as bn
df=pl.read_csv('train.csv')
print(df.describe())
print(df.dtypes)
print(df.null_count())
print(df.mean())
```

```python
#@title количество пассажиров каждого класса
df.get_column('Pclass').value_counts()
```

```python
#@title количество выживших мужчин и женщин на корабле
df.group_by("Sex").agg(pl.col("Survived").sum())
```

```python
#@title часть таблицы с пассажирами, возраст которых больше 44 лет
df.filter(pl.col("Age") > 44)
```

```python
#@title средний возраст пассажиров и его стандартное отклонение
df_pd=pd.read_csv('train.csv')
print('средний возраст пассажиров:', bn.nanmean(df_pd['Age']))
print('стандартное отклонение:', bn.nanstd(df_pd['Age']))
```
